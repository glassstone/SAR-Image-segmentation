```
import cv2
for num in range(1, 37):
    name='tyrol-w'+str(num) 

    img = cv2.imread(valimg_dir + name + ".tif")
    gt = cv2.imread(valgt_dir + name + ".tif")

    height, width, channels = img.shape
    # Number of pieces Horizontally 
    CROP_W_SIZE  = 10 
    # Number of pieces Vertically to each Horizontal  
    CROP_H_SIZE = 10 

    for ih in range(CROP_H_SIZE ):
        for iw in range(CROP_W_SIZE ):
            x = width/CROP_W_SIZE * iw 
            X = int(x)
            y = height/CROP_H_SIZE * ih
            Y = int(y)
            h = (height / CROP_H_SIZE)
            H = int(h)
            w = (width / CROP_W_SIZE )
            W = int(w)
            print(X,Y,H,W)
            img2 = img[Y:Y+H, X:X+W]
            gt2 = gt[Y:Y+H, X:X+W]
            cv2.imwrite("/content/drive/My Drive/sarseg/Data/complete_dataset/crop/images/" + name + str(ih) + str(iw) +  ".tif" ,img2)
            cv2.imwrite("/content/drive/My Drive/sarseg/Data/complete_dataset/crop/gt/" + name + str(ih) + str(iw) +  ".tif" ,gt2)
```
